# visualizing vector field for dipole

```python
import numpy as np 
import matplotlib.pyplot as plt 
  
# Meshgrid 
x, y = np.meshgrid(np.linspace(-5, 5, 10),  
                   np.linspace(-5, 5, 10)) 
  
# vector field for dipole
u = x/(x**2 + y**2) 
v = y/(x**2 + y**2) 

# Plotting Vector Field with QUIVER 
plt.quiver(x, y, u, v, color='red') 
plt.title('Vector Field') 
  
# Setting x, y boundary limits 
plt.xlim(-7, 7) 
plt.ylim(-7, 7) 
  
# Show plot with grid 
plt.grid() 
plt.show() 
```
